```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set(color_codes=True)


import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df.plot(kind="scatter",x ="SepalLengthCm",y ="SepalWidthCm")
```

```python
sns.jointplot(x="SepalLengthCm",y="SepalWidthCm",data=df,size=5)
```

```python
sns.FacetGrid(df,hue="Species",palette="husl",size=5)  \
    .map(plt.scatter,"SepalLengthCm","SepalWidthCm")  \
    .add_legend()
```

```python
sns.FacetGrid(df, hue="Species",palette="hls",size = 5) \
   .map(plt.scatter,"SepalLengthCm","SepalWidthCm") \
   .add_legend()

sns.FacetGrid(df ,hue ="Species",palette="Set2",size = 5) \
   .map(plt.scatter,"SepalLengthCm","SepalWidthCm") \
   .add_legend()

sns.FacetGrid(df ,hue ="Species",palette="cubehelix",size = 5) \
   .map(plt.scatter,"SepalLengthCm","SepalWidthCm") \
   .add_legend()

```

```python
sns.pairplot(df.drop("Id", axis=1), hue = "Species",palette="husl",size=3)
```
